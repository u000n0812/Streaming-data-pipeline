# 1. Introduction

Real-time System은 **Hard**/**Soft**/**Near**/로 분류할 수 있다.  

Hard는 임베디드에서 많이 활용되는데, 시간 요구 사항이 매우 엄격하다. 요구사항이 만족되지 않으면 전체 시스템의 장애 상황으로 판단한다. 여기서는 Hard보다 **Soft**와 **Near**를 더 많이 살펴보자  

Soft와 Near의 개념이 헷갈리면 나중에 혼돈이 발생할 수 있기 때문에 둘중 하나를 결정하는 것은 의외로 중요하다

*데이터가 처리되어 사용될 준비가 되었다면, 이 시점은 Soft realtime인가 아니면 Near realtime인가?*  

**Soft vs Near**를  결정짓는건 쉽지 않다. 그래서 Introduction에서는 Real-time system을 어떻게 정의하고 만들 것인지에 대해 얘기할 것이다.

Data는 **연산 처리 부분**, **클라이언트가 소비하는 부분**으로 나눌 수 있다. 그런데 대부분 시스템은 연산 처리 시스템을 Real-time 방식으로 하지 않는다. 왜냐하면 어차피 클라이언트의 네트워크 이슈, 어플리케이션 설계 영향 등으로 인해 데이터를 소비하지 못할 수도 있고, 이미 어플리케이션이 종료된 상태일 수도 있기 때문이다.  

즉, Hard-Realtime한 방식이 아니라, 클라이언트가 요청하면 그 즉시 꺼내서 쓸 수만 있으면 된다. 이것을 **Streaming Data System**이라고 부른다.

Streaming 방식을 쓰면 Real-time에 대해서는 더이상 걱정할 필요가 없다. 클라이언트가 요청하는 시점에 서버에서 어떤 데이터를 어떻게 전달할 것인지에 집중하면 된다. 이렇게 데이터가 필요할 시점에 전달이 가능한 시스템을 **In-the-moment-system**이라고 부른다.

아키텍처에서 보안은 무엇보다 중요하다. 위에서 *인증*이나 *인가*와 같은 과정이 모두 보안에 해당한다.  

여기서는 보안이 어떤 방식으로 구현되는지 깊이 살펴보지는 않을 예정이다. 다만 각 보안 설정이 어떤 의미를 가지고 있는지는 알아볼 필요가 있다. 보안에 관심이 있다면 *Security Enginerring : A Guide to Building Dependable Distributed Systems, Ross Anderson*을 참고하자.

서비스를 확장하는 방식은 2가지가 있다.  

**수직 확장**은 CPU, 메모리, 하드 드라이브 등의 하드웨어 또는 소프트웨어 리소스를 추가하는 방법이다  

**수평 확장**은 서버에 리소스를 추가하는게 아니라 서버를 추가하는 것이다.  

현대 시스템 디자인에서는 수직 확장을 배제하지는 않지만, 서버의 적정 리소스만 찾고 이후는 수평 확장으로 구현하는 것이 목표이다.
